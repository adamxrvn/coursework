### Group 2

```python
#add group2 columns
items_Group2 = df_delivered.groupby(['Телефон_new']).agg({
    'Группа2': [
        ('КРУПНОГАБАРИТНЫЙ ТОВАР', lambda x: pr(x, 'КРУПНОГАБАРИТНЫЙ ТОВАР')),
        ('ТЕКСТИЛЬ, ТРИКОТАЖ', lambda x: pr(x, 'ТЕКСТИЛЬ, ТРИКОТАЖ')),
        ('ОБУВЬ', lambda x: pr(x, 'ОБУВЬ')),
        ('КАНЦТОВАРЫ, КНИГИ, ДИСКИ', lambda x: pr(x, 'КАНЦТОВАРЫ, КНИГИ, ДИСКИ')),
        ('ТОВАРЫ ДЛЯ КОРМЛЕНИЯ', lambda x: pr(x, 'ТОВАРЫ ДЛЯ КОРМЛЕНИЯ')),
        ('ИГРУШКИ', lambda x: pr(x, 'ИГРУШКИ')),
        ('КОСМЕТИКА/ГИГИЕНА', lambda x: pr(x, 'КОСМЕТИКА/ГИГИЕНА')),
        ('СОПУТСТВУЮЩИЕ ТОВАРЫ', lambda x: pr(x, 'СОПУТСТВУЮЩИЕ ТОВАРЫ')),
        ('ЖЕНСКИЕ ШТУЧКИ', lambda x: pr(x, 'ЖЕНСКИЕ ШТУЧКИ')),
        ('ПОДГУЗНИКИ', lambda x: pr(x, 'ПОДГУЗНИКИ')),
        ('ДЕТСКОЕ ПИТАНИЕ', lambda x: pr(x, 'ДЕТСКОЕ ПИТАНИЕ')),
        ('ТЕХНИКА И ТОВАРЫ ДЛЯ ДОМА', lambda x: pr(x, 'ТЕХНИКА И ТОВАРЫ ДЛЯ ДОМА')),
        ('ТОВАРЫ ДЛЯ ЖИВОТНЫХ', lambda x: pr(x, 'ТОВАРЫ ДЛЯ ЖИВОТНЫХ'))
    ]
})


items_Group2.columns = items_Group2.columns.get_level_values(1) 
items_Group2 = items_Group2.groupby(level = 0).agg('sum');
```

## Final showcase

```python
showcase_Group2 = pd.read_csv('showcase_Group2.csv', sep=';')
showcase_Region = pd.read_csv('showcase_Region.csv', sep=';')
showcase_Type = pd.read_csv('showcase_Type.csv', sep=';')

# Create copies of the original dataframes
showcase1 = showcase_Group2.copy()
showcase2 = showcase_Region.copy()
showcase3 = showcase_Type.copy()

showcase1['sum_Group2'] = showcase1.loc[:, 'КРУПНОГАБАРИТНЫЙ ТОВАР':'ТОВАРЫ ДЛЯ ЖИВОТНЫХ'].sum(axis=1)
showcase2['sum_Region'] = showcase2.loc[:, 'Центральный ф.о.':'Северо-Кавказский ф.о.'].sum(axis=1)
showcase3['sum_Type'] = showcase3.loc[:, 'КГТ':'ИНОЕ'].sum(axis=1)


# Calculate proportions for Group2 columns
col_Group2 = ['КРУПНОГАБАРИТНЫЙ ТОВАР', 'ТЕКСТИЛЬ, ТРИКОТАЖ', 'ОБУВЬ',
       'КАНЦТОВАРЫ, КНИГИ, ДИСКИ', 'ТОВАРЫ ДЛЯ КОРМЛЕНИЯ', 'ИГРУШКИ',
       'КОСМЕТИКА/ГИГИЕНА', 'СОПУТСТВУЮЩИЕ ТОВАРЫ', 'ЖЕНСКИЕ ШТУЧКИ',
       'ПОДГУЗНИКИ', 'ДЕТСКОЕ ПИТАНИЕ', 'ТЕХНИКА И ТОВАРЫ ДЛЯ ДОМА',
       'ТОВАРЫ ДЛЯ ЖИВОТНЫХ']

for col in col_Group2:
    showcase1[col] = (showcase1[col] / showcase1['sum_Group2']).round(2)

# Calculate proportions for Region columns
col_Region = ['Центральный ф.о.', 'Уральский ф.о.', 'Северо-Западный ф.о.', 'Приволжский ф.о.', 'Южный ф.о.', 'Сибирский ф.о.', 'Дальневосточный ф.о.', 'Восточно-Сибирский ф.о.', 'Крымский ф.о.', 'Северо-Кавказский ф.о.']

for col in col_Region:
    showcase2[col] = (showcase2[col] / showcase2['sum_Region']).round(2)

# Calculate proportions for Type columns
col_Type = ['КГТ', 'ОДЕЖДА', 'ИГРУШКИт', 'ППКП', 'ИНОЕ']

for col in col_Type:
    showcase3[col] = (showcase3[col] / showcase3['sum_Type']).round(2)

del showcase1['sum_Group2']
del showcase2['sum_Region']
del showcase3['sum_Type']

# Combine the three dataframes into one
showcase_final = pd.concat([showcase1, showcase2, showcase3], axis=1)

showcase_final.columns
```
